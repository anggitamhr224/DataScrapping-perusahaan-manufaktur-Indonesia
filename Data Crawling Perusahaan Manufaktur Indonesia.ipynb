```python
data_perusahaan = []
data_telepon = []
data_alamat = []
data_deskripsi = []
data_kategori = []

for index in range(1,17):
    page = requests.get(f'https://klikalamat.com/database-daftar-perusahaan-manufaktur-indonesia/page/{index}')
    soup = BeautifulSoup(page.content, 'html.parser')
    
    #mencari nama perusahaan
    perusahaan = soup.find_all("div", class_="listing-title")
    for i in perusahaan:
        a = i.find("h2").text
        data_perusahaan.append(a)
        
    #mencari nomor telepon
    telepon = soup.find_all("p", class_="phone")
    for i in telepon:
        data_telepon.append(i.text)
        
    #mencari alamat
    alamat = soup.find_all("p", class_="address")
    for i in alamat:
        data_alamat.append(i.text)
        if len(data_alamat) == 192:
            break
    
    #mencari deskripsi
    deskripsi = soup.find_all("div", class_="entry-content")
    for i in deskripsi:
        data_deskripsi.append(i.text.strip())
    
    #mencari kategori perusahaan
    kategori = soup.find_all("span", class_="i_category")
    for i in kategori:
        data_kategori.append(i.text[11:])
```

```python
data ={
    "Nama Perusahaan" : data_perusahaan,
    "Telepon" : data_telepon,
    "Alamat" : data_alamat,
    "Deskripsi" : data_deskripsi,
    "Kategori" : data_kategori
}
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(data).to_excel("Database Perusahaan Manufaktur Indonesia.xlsx")
```
